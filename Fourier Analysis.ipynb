# Discrete Fourier Transform on a binary sequence

Code to perform a Discrete Fourier Transform (DFT) on a binary sequence provided by the user, and then display the amplitude and phase of each frequency component in the sequence.

In this Python implementation, we define three functions: read_data to read the binary sequence from the user, dft to perform the Discrete Fourier Transform, and experiment as the main program loop.

To execute the code, you can run the experiment() function, which prompts for the number of data points, reads the binary sequence, performs the DFT, and displays the results.

```python
import numpy as np

def read_data():
    binary_sequence = input("Enter binary sequence (0s and 1s only): ")
    data = np.array([int(digit) for digit in binary_sequence if digit in ['0', '1']])
    return data

def dft(data):
    n = len(data)
    results = np.zeros(n, dtype=complex)
    for k in range(n):
        for j in range(n):
            results[k] += data[j] * np.exp(-2j * np.pi * k * j / n)
    return results

def experiment():
    n = int(input("How many data points to analyze? "))
    data = read_data()
    results = dft(data)
    print("Results:")
    for result in results:
        print(result)

experiment()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

# Generate the time vector
t = np.linspace(0, 2*np.pi, 1000)

# Define the number of harmonics
num_harmonics = 10

# Generate random phase shifts for each harmonic
phase_shifts = np.random.uniform(0, 2*np.pi, num_harmonics)

# Compute the complex waveform
complex_waveform = np.sum([np.exp(1j * (n+1) * t + phase_shift) for n, phase_shift in enumerate(phase_shifts)], axis=0)

# Create the figure and axis
fig, axs = plt.subplots(num_harmonics+1, 1, figsize=(8, 8))
plt.subplots_adjust(left=0.15, bottom=0.35, hspace=0.6)

# Plot the component sine waves
for n, ax in enumerate(axs[:-1]):
    ax.plot(t, np.real(np.exp(1j * (n+1) * t + phase_shifts[n])), lw=1, alpha=0.5)
    ax.set_ylabel(f'Harmonic {n+1}')

# Plot the overlapped waveform
axs[-1].plot(t, np.real(complex_waveform), lw=2, color='black')
axs[-1].set_ylabel('Sum of Harmonics')

# Highlight the original waveform
axs[-1].set_facecolor('lightyellow')

# Create sliders
slider_harmonics = plt.axes([0.15, 0.2, 0.65, 0.03])
slider = Slider(slider_harmonics, 'Number of Harmonics', 1, 50, valinit=num_harmonics, valstep=1)

# Update function when slider is changed
def update(val):
    num_harmonics = int(slider.val)
    phase_shifts = np.random.uniform(0, 2*np.pi, num_harmonics)
    complex_waveform = np.sum([np.exp(1j * (n+1) * t + phase_shift) for n, phase_shift in enumerate(phase_shifts)], axis=0)

    # Update subplots for component sine waves
    for n, ax in enumerate(axs[:-1]):
        ax.clear()
        ax.plot(t, np.real(np.exp(1j * (n+1) * t + phase_shifts[n])), lw=1, alpha=0.5)
        ax.set_ylabel(f'Harmonic {n+1}')

    # Update subplot for overlapped waveform
    axs[-1].clear()
    axs[-1].plot(t, np.real(complex_waveform), lw=2, color='black')
    axs[-1].set_ylabel('Sum of Harmonics')
    axs[-1].set_facecolor('lightyellow')

    fig.canvas.draw_idle()

slider.on_changed(update)

plt.show()
```
