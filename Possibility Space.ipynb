```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Load pretrained word embeddings
# Replace with your own embeddings and vocabulary
word_embeddings = np.random.rand(100, 300)  # 100 words, 300 dimensions
word_vocab = ["Geometric Bayesianism", "Active Inference", "Free Energy Principle", "Perception", "Inference", "Prediction Error", "Neural Dynamics", "Bayesian Brain", "Generative Model", "Markov Blanket", "Precision Weighting", "Sensory Attenuation", "Hierarchical Processing", "Belief Propagation", "Variational Inference", "Complex Systems", "Information Theory", "Neuroscience", "Embodied Cognition", "Self-Organization", "Neural Ensembles", "Epistemic Value", "Homeostasis", "Bayesian Filtering", "Anomalous States", "Perceptual Control", "Active Perception", "Exploration-Exploitation", "Probabilistic Graphical Models", "Bounded Rationality", "Neuronal Firing", "Attentional Modulation", "Action Selection", "Bayesian Optimization", "Neurodynamics", "Neural Coding", "Bayesian Updating", "Predictive Coding", "Multisensory Integration", "Predictive Processing", "Embodied Action", "Model Comparison", "Predictive Coding", "Cognitive Architecture", "Uncertainty Reduction"]

# Apply PCA for dimensionality reduction
pca = PCA(n_components=2)
word_embeddings_2d = pca.fit_transform(word_embeddings)

# Create scatter plot
plt.figure(figsize=(10, 8))
plt.scatter(word_embeddings_2d[:, 0], word_embeddings_2d[:, 1], marker='o')

# Label each point with the corresponding word
for i, word in enumerate(word_vocab):
    plt.annotate(word, (word_embeddings_2d[i, 0], word_embeddings_2d[i, 1]))

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('2D Visualization of Word Embeddings')
plt.show()
```

```python
for i, word in enumerate(word_vocab):
    x, y = word_embeddings_2d[i]
    print(f"{i+1}. {word}: ({x:.2f}, {y:.2f})")
```
